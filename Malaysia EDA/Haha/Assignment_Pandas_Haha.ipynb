## Data Preparation and Cleaning

Data preparation is the important process to clean and transform the data. This step can help to remove missing data, outlier and duplicate value in order to produce quality data analysis results. 



First, we import all the required libraries from Python. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

The data is loaded into the pandas data frame.

```python
df = pd.read_csv('https://raw.githubusercontent.com/peiyu00/data/main/flows-of-foreign-direct-investment-in-malaysia_blo.csv')
df.head()
```

Before preprocessing the data, we should understand the properties of our dataset.

```python
df.info()
```

Describe function is a useful way to know the characteristics of data by getting a summary about the data.

```python
df.describe()
```

We calculate the number of missing value for each column.

```python
print(df.isnull().sum())
```

We can also count the value of not null for each column.

```python
df.notna().sum()
```

Since there are no null values in all of the columns, the dropna() will not affect the size of original data. 

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

We can know the number of duplicated rows for each column and remove them. 

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

However, there are no duplicate rows in our data. So, removing the duplicate row will not affect the size of original data.

```python
df = df.drop_duplicates()
df.count()
```

From the above 4 boxplot, it can be noticed that there exist outliers in the data. Therefore, IQR has been calculated and used to remove outliers. The size of data is decreased to 1287.

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

## Asking and Answering Questions

A total of 6 questions will be asked in this section and all of them were answered by using pandas, series or visualization(matplotlib & seaborn).



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Q1: Which quarter of every year is profiting?

We can observe the profit quarter by looking at the “Net RM million” column. Pandas groupby function has been used to group the data based on year and quarter. Then, the sum function computes the sum of numeric columns based on year and quarter. 

```python
data = df.groupby(['Year', 'Quarter']).sum()
data
```

A line chart of “Net RM million” against “Quarter” for each year was plotted by using seaborn library(lineplot) and the label the legend by using the library from matplotlib. 

```python
for i in range(2008, 2023):
  sns.lineplot(data=df[df['Year']==i], x='Quarter',y='Net RM million', ci=None, label = i)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

**Answer:** By observing both dataframe and line chart, there is only the second quarter of 2009 with negative value.Therefore, all of the quarters in every year are profitable except second quarter. 

#### Q2: Which country has the most confidence in Malaysia?

The confidence of other countries toward Malaysia can be measured by the amount of funds that invested by that country. In this case, we can get the data by grouping the data(pandas library) based on countries and calculate the sum of “Credit RM million“. Then, sort the data by using the sort_values() function from pandas library in order to make the observation easier. 

```python
data = df.groupby('Countries')
max = data['Credit RM million'].sum().sort_values(ascending = False)
max
```

A bar chart from seaborn library has been used to plot the top 5 countries that have the largest credit of funds to Malaysia. 

```python
plt.figure(figsize=(20, 9))
bar_chart = sns.barplot(max.index[1:6], max.values[1:6])
for bar in bar_chart.patches:
  bar_chart.annotate(bar.get_height(),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=9, xytext=(0, 8),
                   textcoords='offset points')
plt.show()
```

**Answer:** Singapore has the most confident on Malaysia, followed by Japan. 

#### Q3: Which block of country brings the positive effect to Malaysia in terms of economy?

We want to investigate which blocks of countries profit from the investment in Malaysia. We can get the data by grouping the data based on blocks of countries(pandas library) and compute the “Net RM million”. 

```python
country = df.groupby('Blocks of countries').sum()
country
```

To obtain a better view, we plot the pie chart by using matplotlib library function (pie) and label the legends based on blocks of countries. 

```python
plt.pie(country['Net RM million'], labels=country['Net RM million'])
plt.legend(country.index, loc='right', bbox_to_anchor=(2,0.5))
plt.show()
```

**Answer:** According to both dataframe and pie chart, all blocks of countries brought the positive effect to the economy of Malaysia. 

#### Q4: Which year is the most profitable?

We can observe the profit by looking at the “Net RM million” column. Pandas groupby function has been used to group the data based on year Then, the sum function computes the sum of "Net RM million" columns based on year. 

```python
data2 = df.groupby('Year')
max2 = data2['Net RM million'].sum().sort_values(ascending = False)
max2
```

A bar chart(bar()) from matplotlib library has been used to plot the "Year" vs "Net RM million".

```python
plt.bar(max2.index, max2.values)
plt.show()
```

**Answer:** By observing both series and bar chart, 2012 is the most profitable year.

#### Q5: How many countries for each block of countries are considered as a main investor in Malaysia?

The number of invested countries in each block of countries can be obtained by grouping the data(groupby) based on blocks of countries first. Then, nunique() function of the pandas library will be used to calculate the number of distinct countries in each block of countries. 

```python
data2 = df.groupby('Blocks of countries')['Countries'].nunique()
data2
```

The horizontal bar chart was plotted by using the matplotlib library function(barh) and set the title(title), x axis label(xlabel) and y axis label(ylabel) also. 

```python
data2.plot.barh()

plt.title('Number of Countries in Each Blocks of Countries')
plt.ylabel('Number of countries')
plt.xlabel('Blocks of countries')
plt.show()
```

**Answer:** There are 6 countries for East Asia, 8 countries for Europe, 4 countries for Latin America, 3 countries for Southeast Asia, 2 countries for North America and Oceania are considered as the investors in Malaysia. 

#### Q6: Which of the top 3 countries have the highest impact toward Malaysia investment during all these years?

We had used 'groupby' function to group up based on the countries names and 'sum()' to sum up each of their “Net RM million” column based on countries name in order to indicate the most impactful country. 

```python
data1 = df.groupby('Countries')
max1 = data1['Net RM million'].sum().sort_values(ascending = False)
m = max1[1:]
m
```

A histogram graph about sum of amount “Net RM million”  countries' name.

```python
plt.title('Countries with highest Impact')
plt.bar(m.index, m.values)
plt.xticks(rotation=90)
plt.show()
```

**Answer:** Based on the graph shown above, the highest is Japan, followed by United States and the third is Singapore.
