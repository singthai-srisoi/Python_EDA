# Data Preparation and Cleaning

**Data** **cleaning** is the process of modifying data to ensure that it is free of irrelevances and incorrect information. It entails identifying **incorrect**, **irrelevant**, **incomplete**, and the **dirty** parts of a dataset and then replacing or cleaning the dirty parts of the data. 

It is very valuable in improving the efficiency of the result of data analysis as it helps to improve data quality.

The process of data cleansing may involve the removal of typographical errors, data validation, and data enhancement. This will be done until the data is reported to meet the data quality criteria, which include; **validity, accuracy, completeness, consistency, and uniformity**.



- Below, we can see the RangeIndex of each column which is 1646.
- In the dataset, there is 1 column that had float64 as data type while 2 are int64 and the rest are object.
- Aslo, we can see this dataset has no null values.

```python
election.info()
```

- First, we remove the **%** symbol for **% of total Votes** column
- And, we convert the columns to a suitable data type

```python
election = election.replace('%','')
election[['State','Seat ID','Seat Name','Candidate Name','Candidate Party','% of total Votes',"Gender",'Status','Pekerjaan']] = election[['State','Seat ID','Seat Name','Candidate Name','Candidate Party','% of total Votes',"Gender",'Status','Pekerjaan']].convert_dtypes(str)

election['Votes for Candidate'] = election['Votes for Candidate'].convert_dtypes(int)
election.info()
```

- Then, we replace 0 in **Pekerjaan** column to **TIADA** to standardize the data.


```python
election['Pekerjaan'] = election['Pekerjaan'].replace('0','TIADA')
election
```

## **Asking and Answering Questions**



#### **Question 1:** Which state have cast the highest amount of votes?





```python
vote_state = election.groupby('State')['Total Votes Cast'].sum()
vote_state.plot(kind = 'bar', color='skyblue')
```

> **Answer:** Selangor is the state that has the highest amount of votes during 2018's general election followed by Johor and Sabah. 

#### **Question 2:** Are the majority of the winning candidates male or female?




```python
election['Status'].unique() 
```

```python
election['Status'].value_counts()
```

```python
fig = px.bar(election, x='Status', color='Gender', facet_col='Gender')
fig.show("svg", engine="kaleido")
```

> **Answer:** The majority of winning candidates are Male (L) which is more than 400 candidates meanwhile the winning candidates for Female(P) are only below than 100 candidates.

#### **Question 3**: Which state(s) has big difference between male and female in terms of total votes casted?




```python
sns.catplot(data=election, kind="violin", x="State", y="Total Votes Cast", hue="Gender", split=True, height=6, aspect=6)
```

> From the catplot above, Melaka is one of the states that have big range of differences between male and female in terms of total votes cast. We can assume that this result is either because there being only a few female candidates competed or the voters put their trust and place the responsibilities on the majority of the male candidates.

#### **Question 4:** Which party won the highest amount of seats during Malaysia's 14th election?


```python
fig = px.pie(election, values="Candidate Win", names="Candidate Party", hover_name="Candidate Party", labels={'Candidate Party':'CandidateWin'}, title="Number of seats per party")
fig.update_traces(textposition='inside', textinfo='percent+label+value',marker=dict(line=dict(color='#000000', width=2)))
fig.show("svg", engine="kaleido")
```

> **Answer:** PKR won the highest amount of seats among other parties during Malaysia's 14th general election which 43.6%.
