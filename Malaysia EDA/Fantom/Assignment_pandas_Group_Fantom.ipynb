From the above 10 boxplot, most of the outliers are exist in the right side of our data. 
Hence, we need to calculate the IQR and remove the unwanted outliers.

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

#### Q1: Which state have greatest death rate?

We can observe the death rate by dividing the total new case with the death case to obtain the death rate. Pandas groupby function will be used to group the data based on the state.<br><br>
The Covid death data will be obtain by using new dataset and merging then to provide the death rate output.

Firstly import the new dataset

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/MoH-Malaysia/covid19-public/main/epidemic/deaths_state.csv")
df2.head()
```

```python
df['Total case'] = 0
for features in df[['cases_new','cases_import','cases_active','cases_cluster','cases_unvax','cases_pvax','cases_fvax','cases_boost','cases_child','cases_adolescent','cases_adult','cases_elderly','cases_0_4','cases_5_11','cases_12_17','cases_18_29','cases_30_39','cases_40_49','cases_50_59','cases_60_69','cases_70_79','cases_80']]:
  df['Total case'] = df['Total case'] + df[features]
data_new_cases = df.groupby('state')['Total case'].sum()
data_new_cases
```

After getting the total case of each state we will now obtain the total death of each states. 

```python
df2['Total death'] = 0
for features in df2[['deaths_new','deaths_bid']]:
  df2['Total death'] = df2['Total death'] + df2[features]
data_death_cases = df2.groupby('state')['Total death'].sum()
data_death_cases
```

Now we will the calculate the death rate for each states.

```python
new_data = pd.concat([data_new_cases, data_death_cases], axis = 1)
new_data
```

```python
new_data['Death rate'] = (new_data['Total death']/new_data['Total case'])*100
new_data
```

```python
axes = new_data.plot.bar(rot=0, subplots=True, figsize=(25,15))
axes[1].legend(loc=2)  
```

**Answer:** By observing the dataframe and barchart, it is obvious that the selangor have the highest death rate compare to other states. 

#### Q2: Which state have the highest recovery rate. 

The recovery rate can be measured by the amount of cases recovered divide by the total cases of the state. Pandas groupby function will be used to group to state and sort_values() function will be used to make observation earsier.

Firstly we obtain the total recovery cases and total cases of each state.

```python
data = df.groupby('state')[['cases_recovered', 'Total case']].sum()
data
```

Next we calculate the Recovery rate by using the formula below.

```python
data['Recovery rate'] = (data['cases_recovered']/data['Total case'])*100
data
```

Then we sort the recovery rate using sort_values() function.

```python
data_recovery = data['Recovery rate'].sort_values(ascending = False)
data_recovery
```

Lastly, we will plot a bar chart to provide better visualization on our output.For this bar chart we will display top 5 states that have highest recovery rate in Malaysia.

```python
plt.bar(data_recovery.index[0:5], data_recovery.values[0:5])
plt.show()
```

#### Q3: What is the recovered cases across the states?

```python
state = df.groupby('state').sum()
state
```

```python
plt.pie(state['cases_recovered'], labels=state['cases_recovered'])
plt.legend(state.index, loc='right', bbox_to_anchor=(2,0.5))
plt.show()
```

**Answer:** According to pie chart, Selangor state had the highest recovered cases across the states.
